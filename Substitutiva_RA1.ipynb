### Exercício 1
$$ \lim_{x \to 0} \frac{\sqrt{x+9} - 3}{x} $$

Começamos substituindo o  $x$ pelo limite, chegando no resultado:


$$ \lim_{x \to 0} \frac{\sqrt{x+9} - 3}{x} = \frac{\sqrt{0+9} - 3}{0} = \frac{\sqrt{9} - 3}{0} = \frac{3 - 3}{0} = \frac {0}{0} $$

Como $\frac {0}{0}$ é indeterminado, podemos aplicar os princípios da radiciação, multiplicando o numerador e o denominador pelo conjugado do numerador, chegando em:

$$\lim_{x \to 0} \frac{\sqrt{x+9} - 3}{x}⋅\frac{\sqrt{x+9} + 3}{\sqrt{x+9} + 3} = \frac{(\sqrt{x+9})^2 - 3^2}{x ⋅ \sqrt{x+9} + 3}$$
$$\frac{{x+9} - 9}{x ⋅ (\sqrt{x+9} + 3)} = \frac{{x}}{x ⋅ (\sqrt{x+9} + 3)}$$

Eliminando o $x$ do numerador com o $x$ do denominador, chegamos em:
$$\frac{{1}}{\sqrt{x+9} + 3}$$

Por fim, substituindo x pelo limite, o resultado fica:
$$\lim_{x \to 0} \frac{{1}}{\sqrt{0+9} + 3} = \frac{{1}}{\sqrt{9} + 3} =\frac{{1}}{3 + 3} = \frac{1} {6}$$



```python
import matplotlib.pyplot as plt
import numpy as np
import math

# Define a função
def f(x):
    return math.sqrt(9 + 3) / x

# Define os valores de x
x_values = np.linspace(0.1, 2, 100)  # valores de 0.1 a 2

# Calcula os valores de y
y_values = [f(x) for x in x_values]

# Plota o gráfico da função
plt.plot(x_values, y_values, label=r'$ \frac{\sqrt{x+9} - 3}{x}$')



# Adiciona legendas aos eixos
plt.xlabel('x')
plt.ylabel('f(x)')

# Adiciona título ao gráfico
plt.title('Gráfico da função')

# Adiciona a legenda
plt.legend()

# Exibe o gráfico
plt.grid(True)
plt.show()
```

### Exercício 2
$$ \lim_{x \to \infty} \frac{25x + 7}{\sqrt{5x^2 + 2}} $$

Dividindo os números da equação por $x$, nós temos:
$$\frac{\frac{25x}{x} + \frac {7} {x}}{\sqrt{\frac{5x^2} {x^2} + \frac {2} {x^2}}}$$

Simplificando, ficamos com:
$$\frac{25 + \frac{7}{x}} {\sqrt{5+ \frac {2} {x^2}}}$$

A medida em que $x$ tende ao infinito, o termo $\frac{7} {x} $ também se aproxima de 0. Então a expressão dentro da raiz se aproxima de $\sqrt 5$, pois o termo constante 5 é dominante

Então temos:

$$\frac{25x + 7}{\sqrt{5x^2 + 2}} = \frac {25} {\sqrt 5} = \frac {25 \sqrt 5} {5} = 5 \sqrt5$$

```python
import matplotlib.pyplot as plt
import numpy as np
import math

# Define a função
def f(x):
    return (25*x + 7) / math.sqrt(5*x**2 + 2)

# Define os valores de x
x_values = np.linspace(1, 10, 100)  # valores de 1 a 10

# Calcula os valores de y
y_values = [f(x) for x in x_values]

# Plota o gráfico da função
plt.plot(x_values, y_values, label=r'$\frac{25x + 7}{\sqrt{5x^2 + 2}} $')

# Adiciona uma linha indicando o limite quando x tende ao infinito


# Adiciona legendas aos eixos
plt.xlabel('x')
plt.ylabel('f(x)')

# Adiciona título ao gráfico
plt.title('Gráfico da função ')

# Adiciona a legenda
plt.legend()

# Exibe o gráfico
plt.grid(True)
plt.show()
```
